# Employ a takeaway title

The original title was not clear with the data of each lego sets was sold each year. Therefore, I changed the original title to Star Wars is The Most Sold Lego Sets. Also, I make the star Wars sets apeared on the top of the chart to make it more clear that it is sold the most.

```python

url_lego = 'https://raw.githubusercontent.com/3eltran23/InfoViz/master/LegoTotalSetsPerYear.csv'

# reading dataframe
df_lego = pd.read_csv(url_lego)
df_lego = df_lego[df_lego['year']> 2017]
df_lego.head(11)
```

```python
labels = ['Others', 'Harry Potter', 'Marvel','Star Wars']
qcs =['#BFBEBE','#7570b3','#d95f02','#1b9e77']
width = 0.6 

bottom = np.zeros((len(df_lego),), dtype=int)  # bottom accumulates starting points


fig, ax = plt.subplots(figsize=(15,8))

for pos, label in enumerate(labels):
  
    ax.bar(df_lego['year'], df_lego[label], width=width, label=label, color=qcs[pos], bottom=bottom,edgecolor='white')
    bottom = bottom +  df_lego[label]
 
ax.set_ylabel('LEGOs sold')
ax.set_xlabel('Years')
ax.legend()
ax.set_title('Star Wars Is the Most Sold Lego Sets', fontsize=26, loc='left')

plt.show()
```

# Label data Directly

In the graph I added two labels for x and y axis to make the view more striaghtforward. Also I added the sum of each column on top of them to help useres to get the data more directly.

```python
labels = ['Others', 'Harry Potter', 'Marvel','Star Wars']
qcs =['#BFBEBE','#7570b3','#d95f02','#1b9e77']
width = 0.6 

bottom = np.zeros((len(df_lego),), dtype=int)  # bottom accumulates starting points


fig, ax = plt.subplots(figsize=(15,10))

for pos, label in enumerate(labels):
  
    ax.bar(df_lego['year'], df_lego[label], width=width, label=label, color=qcs[pos], bottom=bottom,edgecolor='white')
    bottom = bottom +  df_lego[label]

totals = df_lego.sum(axis=1)

# Set an offset that is used to bump the label up a bit above the bar.
y_offset = 4
# Add labels to each bar.
year = [2018, 2019, 2020, 2021, 2022]

for i, total in enumerate(totals):
    print(total - year[i] + y_offset)
    ax.text(year[i], total-year[i] + y_offset,round(total-year[i]), ha='center',
          weight='bold')
ax.set_ylabel('LEGO sets')
ax.set_xlabel('Year')
ax.legend()
ax.set_title('Star Wars is The Most Sold Lego Sets', fontsize=26, loc='left')

plt.show()
```

# Use sufficient contrast

I used coblis and color contrast analyzer for this chart and I found our the previous graph is not so straightforward for color blind. Therefore I changed new color to it.
